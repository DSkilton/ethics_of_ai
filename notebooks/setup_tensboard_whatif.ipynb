```python
import os
import sys
import datetime

import pandas as pd
import tensorflow as tf
import torch

print("Python executable:", sys.executable)
print("TensorFlow:", tf.__version__)
print("TF GPUs:", tf.config.list_physical_devices('GPU'))
print("PyTorch:", torch.__version__)
print("Torch CUDA available:", torch.cuda.is_available())
print("Torch device:", torch.cuda.get_device_name(0) if torch.cuda.is_available() else "no gpu")

DATA_PATH = "../data/thyroid_cancer_risk_data.csv"
df = pd.read_csv(DATA_PATH)

print("Shape:", df.shape)
print(df.head())
print(df.dtypes)
```

```python
%load_ext tensorboard
```

```python
# TODO: delete this test data
model_name = "shared_expert_mlp"
use_class_weights = True
use_fairness_loss = False
num_experts = 4
dropout = 0.2
learning_rate = 0.001
batch_size = 64
seed = 42
timestamp = "20260404_1530"
```

```python

run_name = (
    f"{model_name}"
    f"_cw{int(use_class_weights)}"
    f"_fair{int(use_fairness_loss)}"
    f"_experts{num_experts}"
    f"_drop{dropout}"
    f"_lr{learning_rate}"
    f"_bs{batch_size}"
    f"_seed{seed}"
    f"_{timestamp}"
)

log_dir = os.path.join("../logs", "fit", run_name)

os.makedirs(log_dir, exist_ok=True)
print("Log dir:", log_dir)
```

```python
writer = tf.summary.create_file_writer(log_dir)

with writer.as_default():
    for step in range(5):
        tf.summary.scalar("test_metric", step * 0.1, step=step)

print("Test TensorBoard logs written.")
```
